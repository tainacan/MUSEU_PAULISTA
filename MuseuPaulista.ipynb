```python
#Informações de dbfread em https://readthedocs.org/projects/dbfread/downloads/pdf/latest/
import dbfread
import pandas as pd
import numpy as np

```

```python
#The file name is case insensitive, which means DBF(’PEOPLE.DBF’) will open the file people.dbf. If
#there is a memo file, it too will be looked for in a case insensitive manner, so DBF(’PEOPLE.DBF’) would
#find the memo file people.FPT.
#DBFNotFound will be raised if the file is not found, and MissingMemoFile if the memo file is missing.

#Precisa ter os arquivos .DBF e .FPT(memo file) 
table = dbfread.DBF('C:/Users/carol/Desktop/projeto_museu_paulista/Banco dados original/objetos.dbf', load=True)

#Transforma em dataframe pandas
frame = pd.DataFrame(iter(table))
frame
```
